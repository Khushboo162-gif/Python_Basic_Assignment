Ques 1: What is the name of the feature responsible for generating Regex objects?

Ans 1: The re.compile() function returns Regex objects.

Ques 2: Why do raw strings often appear in Regex objects?

Ans 2: Raw strings are used so that backslashes do not have to be escaped.

Ques 3: What is the return value of the search() method?

Ans 3: The search() method returns Match objects.

Ques 4: From a Match item, how do you get the actual strings that match the pattern?

Ans 4: The group() method returns strings of the matched text.

Ques 5: In the regrex which created from the r'(\d\d\d)-(\d\d\d-\d\d\d\d)', what does group zero cover? Group 2? Group 1?

Ans 5: Group 0 is the entire match, group 1 covers the first set of parentheses, and group 2 covers the second set of parentheses.

Ques 6: In standard expression syntax,parantheses and intervals have distinct meanings. How can you tell a regrex that you want it to fit real parantheses and periods?

Ans 6: Periods and parentheses can be with a backslash: \ . , \ (, and \ ).

Ques 7: The findall() method returns a string list or a list of string tuples. What cause it to return one of the two options?

Ans 7: If the regex has no groups,a list of strings is returned. If the regex has groups, a list of tuples of strings is returned.

Ques 8: In standard expressions, what does the | character mean?

Ans 8: The | character signifies metching "either, or" between two groups.

Ques 9: In regular expressions, what does the character stand for?

Ans 9: The ? charactercan either mean "match zero or one of the pecending group" or be used to signify nongreedy matching.

Ques 10: In regular expressions, what does the difference between the + and * characters?

Ans 10: The + matches one or more. The * matches zero or more.

Ques 11: What is the difference between {4} and {4,5} in regular expressions?

Ans 11: The {4} matches exactly three instances of the precending group. The {4,5} matches between three and five instances.

Ques 12: What do you mean by the \d,\w and \s shorthand character classes signify in regular expressions?

Ans 12: The \d, \w, and \s shorthand character classes match a single digit, word, or space character, respectively.

Ques 13: What do means by \D,\W and \S shorthand character classes signify in regular expressions?

Ans 13: The \D, \W, and \S shorthand character classes match a single character that is not a digit, word, or space character, respectively.

Ques 14: What is the difference between .* ? and .* ?

Ans 14: The . performs a greedy match, and the .? performs a nongreedy match.

Ques 15: What is the syntax for matching both numbers and lowercase letters with a character class?

Ans 15: Either [0-9a-z] or [a-z0-9]

Ques 16: What is the procedure for making a normal expressions in regrex case insensitive?

Ans 16: Passing re.l or re.IGNORECASE as the second argument to re.complie() will make the matching case insensitive.

Ques 17: What does the . character normally match? What does it match if re.DOTALL is passed 2nd argument in re.compile()?

Ans 17: The . character mormally matches any character except the newline character. If re.DOTALL is passed as the second argument to the re.complie(), then the dot will also match newline characters.

Ques 18: If numReg = re.compile(r'\d+'), what will numRegex.sub('X','11 drummers, 10 pipers, five rings, 4 hen') return?

Ans 18: 'X drummers, X pipers, five rings, X hens'

Ques 19: What does passing re.VERBOSE as the 2nd argument to re.compile() allow to do?

Ans 19: The re.VERBOSE argument allows you to add whitespace and comments to the string passed to re.complie().

Ques 20: How would you write a regex that match a number with comma for every three digits? It muct match the given following:
        
        '42'
        '1,234'
        '6,368,745'
        
        but not the following:
            
        '12,34,567'(which has only two digits between the commas)
        '1234'(which lacks commas)

Ans 20: re.compile(r'^\d{1,3}(,\d{3})*$') will create this regex, but other regex strings can produce a similar regular expression.

Ques 21: How would you write a regex that matches the full name of someone whose last name is Watanabe? You can assume that the first name that comes before it will always be one word that begins with a capital letter. The regex must match the following:
        
        'Haruto Watanabe'
        'Alice Watanabe'
        'RoboCop Watanabe'
        but not the following:
        'haruto Watanabe'(where the first name is not capitalized)
        'Mr.Watanabe'(where the preceding word has a nonletter character)
        'Watanabe'(which has no first name)
        'Haruto watanabe'(where Watanabe is not capitalized)

Ans 21: re.compile(r'[A-Z][a-z]*\s Watanabe')
