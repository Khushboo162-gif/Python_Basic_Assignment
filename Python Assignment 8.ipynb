Ques 1: Is the Python Standard Library included with PyInputPlus?

Ans 1: No, PyInputPlus is not part of python standard library. It is installed using pip, pip install PyInputPlus.

Ques 2: Why is PyInputPlus commonly imported with import pyinputplus as pypi?

Ans 2: So that we can enter a shorter name when calling the module's function.

Ques 3: How do you distinguish between inputInt() and inputFloat()?

Ans 3: inputInt() accepts an integer value, inputFloat() accepts a floating-point numeric value.

Ques 4: Using PyInputPlus, how do you ensure that the user enters a whole number between 0 and 99?

Ans 4: We have to assign min=0 and max=99 inside pyinputplus.inputInt().

Ques 5: What is transferred to the keyboard arguments allowRegexes and blockRegexes?

Ans 5: allowRegexes and blockRegexes keyword take a list of regular expression as a string to determine what pyinputplus function will accept or reject as valid input.

Ques 6: If a blank input is entered three times, what does inputStr(limit=3) do?

Ans 6: ValidationException and RetryLimitException will be returned

Ques 7: If blank input is entered three times, what does inputStr(limit=3,default='hello') do?

Ans 7: It returns 'hello'.
