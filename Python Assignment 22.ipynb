2: What is the result of the code, and explain?

```python
X = 'iNeuron'
def func():
    X = 'NI'
func()
print(X)
```

print(X) prints 'iNeuron' which is global variable, inside func() X='NI' which is local variable

6: How about this code: what is its output in Python 3, and explain?

```python
def func():
    X = 'NI'
    def nested():
        nonlocal X
        X = 'Spam'
    nested()
    print(X)
func()
```
