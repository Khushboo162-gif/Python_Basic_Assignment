Ques 1: Why are functions advantageous to have in your program ?

Ans 1: Functions reduce the need for duplicate code. This makes programs shorter, easier to read, and easier to update.

Ques 2: When does the code in a function run: when it's specified or when it's called ?

Ans 2: The code in a function executes when the function is called, not when the function is defined.

Ques 3: What statement creates a function ?

Ans 3: The def statement defines(that is, creates) a function.

Ques 4: What is the difference between a function and a function call ?

Ans 4: Afunction consists of the def statement and the code in its def clause.
    
    A function is what moves the program execution into the function, and the function call evaluatesto the function's return value.

Ques 5: How many global scopes are there in a Python program ? How many local scopes?

Ans 5: There is one global scope, and a local scope is created whenever a function is called.

Ques 6: What happens to variables in a local scope when the function call returns?

Ans 6: When a function returns, the local scope is destroyed, and all the variables in it are forgotten.

Ques 7: What is the concept of a return value? Is it possible to have a return value in an expression?

Ans 7: A return value is the value that a function call evaluates to. Like any value, a return value can be used as part of an expression.

Ques 8: If a function does not have a return statement, what is the return value of a call to that function?

Ans 8: If there is no return statement for a function, its return value is None.

Ques 9: How do you make a functio variable refer to the global variable?

Ans 9: A global statement will force a variable in a function to refer to the global variable.

Ques 10: What is the datatype of None?

Ans 10: The data type of None is NoneType.

Ques 12: If you had a bacon() feature in a spam module, what would you call it after importing spam?

Ans 12: This function can be called with spam.bacon().

Ques 13: What can you do to save a program from crashing if it encounters an error?

Ans 13: Place the line of code that might cause an error in a try clause.

Ques 14: What is the purpose of the try clause? What is the purpose of the except clause?

Ans 14: The code that could potentially cause an error goes in the try clause.
    
    The code that executes if an error happens goes in the except clause.
