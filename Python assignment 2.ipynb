QUES5:- What are the six comparison operators?

ANS5:-(i) Less Than(<) Operator: it checks if the left value is lesser than that on the right.
       
       eg: 3 < 6 #displays True
(ii)Greater Than(>) Operator: it checks whether the left value is greater than the one on the right.

eg: 0.5 > False #displays True

(iii) Less Than or Equal To(<=) Operator: it returns True only if the value on the left is either less than or equal to that on the right of the operator.

eg: a = 2
    
    a<=a*2 #displays True

(iv) Equal To or Greater Than(>=) Operator: it returns True only if the value on the left is greater than or equal to that on the right.

eg: from math import pi

    3.14>=pi #displays False

(v) Equal To(==) Operator: it returns True if the values on either side of the operator are equal.

eg: 3 == '3' #displays False

(vi) Not Equal (!=) Operator: it returns True if the values on either side of the operator are unequal.

eg: 3 != 3.0 #displays False
