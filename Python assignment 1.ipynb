Ques4:- What is an expression made up of? What do all expressions do?

Ans4:- An expression is a combination of operators and operands that is interpreted to produce some other value.An expression is evaluted as per the precedence of its operators.We have many different types of expressions in Python.

1.CONSTANT EXPRESSIONS:- These are the expressions that have constant values only.

eg: x = 15 + 1.3

    print(x) #diplays 16.3
    
2.ARITHMETIC EXPRESSIONS:- An arithmetic expression is a combination of numeric values, operators, and sometimes parenthesis.

eg: x = 40
    
    y = 12
    
    add = x+y
    
    sub = x-y
    
    pro = x*y
    
    div = x/y
    
    print(add) #displays 52
    
    print(sub) #displays 28
    
    print(pro) #displays 480
    
    print(div) #displays 3.333333333
    
3.INTEGRAL EXPRESSIONS:- These are the kind of expressions that produce only integer results after all computations and type conversions.

eg: a = 13
    
    b = 12.0
    
    c = a + int(b)
    
    print(c) #displays 25
    
4. FLOATING EXPRESSIONS:- These are the kind of expressions which poroduce floating point numbers as result after all computations and type conversions.

eg: a = 13
    
    b = 5
    
    c = a / b
    
    print(c) #displays 2.6
    
5.RELATIONAL EXPRESSIONS:- In these type of expressions, arithmetic expressions are written on both sides of relational operator(> , < , >= , <=).

eg: a = 21
    
    b = 13
    
    c = 40
    
    d = 37
    
    p = (a + b) >= (c - d)
    
    print(p) #displays True
    
6.LOGICAL EXPRESSIONS:- These are kinds of expressions that result in either True or False.

eg: p = (10 == 9)
    
    q =(7 > 5)
    
    R = p and q
    
    S = p or q
    
    T = not p
    
    print(R) #displays False
    
    print(S) #displays True
    
    print(T) #displays True
    
7. BITWISE EXPRESSIONS:-These are the kind of expressions in which computations are performed at bit level.

eg: a = 12
    
    x = a >> 2
    
    y = a << 1
    
    print(x , y) #displays 3 24
    
8.COMBINATIONAL EXPRESSIONS:- We can also use different  types of expressions in a single epressions, and that will be termed as combinational expressions.

eg: a = 16
    
    b = 12
    
    c = a + (b >> 1)
