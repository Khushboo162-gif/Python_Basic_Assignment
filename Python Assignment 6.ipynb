Ques 1: What are escape characters, and how do you use them?

Ans 1: Escape characters represents characters in string values that would otherwise be difficult or impossible to type into code.

Ques 2: What do the escape character n and t stand for?

Ans 2: \n is a newline; \t is a tab

Ques 3: What is the way to include backslash characters in string?

Ans 3: The \\ escape character will represent a backslash character.

Ques 4: The string "Howl's Moving Castle" is a correct value. Why isn't the single quote character in the word Howl's not escaped a problem?

Ans 4: The single quote in Howl's is fine because you've used double quotes to mark the beginning and end of the string.

Ques 5: How do you write a string of newlines if you don't want to use the n character?

Ans 5: Multiline strings allow you to use newlines in strings without the \n escape character.

Ques 9: What are the methods for right-justifying,left-justfying and centering a string?

Ans 9: The rjust(),ljust(),and center() string methods, respectively.

Ques 10: What is the best way to remove whitespace characters from the start or end?

Ans 10: The lstrip() and rstrip() methods remove whitespace from the left and right of a string, respectively.
