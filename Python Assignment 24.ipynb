Ques 5: What are recusive functions, and how are they used?

Ans 5: When a function call itself the this kind of function is called recursive function.

```python
def factorial(n):
    if n==0 or n==1:
        return 1
    else:
        fact=n*factorial(n-1)
        return fact
```

Here factorial() is a recursive function.
