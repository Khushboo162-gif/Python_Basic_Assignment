Ques 3: What is a box tuple, and how does it work?

Ans 3: Many Pillow's function take box tuple argument. This means pillow is taking a tuple of four integer coordinates that expecting a rectangular region in image. The four integers are in order left, top, right, bottom.

Ques 4: Use your image and load in notebook then, how can you find out the width and height of an image object?

```python
from PIL import Image
im=Image.open("image.jpg")
width, height = im.size
print("width=", width)
print("height=", height)
```
