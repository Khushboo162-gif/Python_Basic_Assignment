Ques 1: In what modes should the PdfFileReader() and PdfFileWriter() File objects will be opened?

Ans 1: Read-binary('rb') for PdfFileReader() and write-binary('wb') for PdfFileWriter().

Ques 2: From a PdfFileReader object, how do you get a Page object for page 5?

Ans 2: Calling getPage(4) will return a Page object for about this book, since page 0 is the first page.

Ques 3: What PdfFileReader variable stores the number of pages in the PDF document?

Ans 3: The numPages variable stores an integer of the number of pages in the PdfFileReader object.

Ques 4: If a PdfFileReader object's PDF encrypted with the password swordfish, what must you do before you can obtain Page objects from it?

Ans 4: Call decrypt('swordfish')

Ques 5: What methods do you use to rotate a page?

Ans 5: rotateClockwise() and rotateCounterClockwise()

Ques 6: What is the difference between a Run object and a Paragraph object?

Ans 6: A document contains multiple paragraphs. A paragraph begins on a new line and contains multiple runs. Runs are contiguous groups of characters within a paragraph.

Ques 7: How do you obtain a list of Paragraph objects for a Document object that's stored in a variable named doc?

Ans 7: doc.paragraphs

Ques 8: What type of object has bold, underline, italic, strike, and outline variables?

Ans 8: runs

Ques 9: What is the difference between False, True, and None for the bold variable?

Ans 9: True = bold is always enabled no matter what other styles are applied to the run, False = bold is always disabled, None = defaults to whatever the run's style set to

Ques 10: How do you create a Document object for a new Word document?

Ans 10: Call the docx.Document() function.

Ques 11: How do you add a paragraph with the text 'Hello, there!' to a Document object stored in a variable named doc?

Ans 11: doc.add_paragraph('Hello there!')

Ques 12: What integers represent the levels of headings available in Word documents?

Ans 12: 0 to 4
